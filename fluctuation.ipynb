```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#dataset
data = {
    "Age": [25, 34, 41, 29, 50, 31, 45, 22, 37, 28],
    "Gender": ["M", "F", "M", "F", "M", "F", "F", "M", "F", "M"],
    "Department": ["IT", "HR", "IT", "Sales", "IT", "Sales", "HR", "IT", "Sales", "IT"],
    "Salary": [1500, 1300, 2000, 1600, 2300, 1500, 1800, 1900, 2100, 1500],
    "Length": [1, 3, 7, 5, 10, 5, 2, 7, 0.5, 6],
    "Left": [1, 0, 0, 1, 0, 1, 0, 0, 0, 1]
}
```

## ðŸ“ŠDescriptive statistics

The dataset provides basic statistical information.
The average age of employees is around 34 years.
The average salary is approximately â‚¬1750.
Employees' length in the Company ranges from 0.5 to 10 years.
There is a noticeable spread in the data, showing diversity among employees.

```python
df = pd.DataFrame(data)
print(df.describe().round(2))
```

## ðŸ“Š Age Distribution

This histogram shows how employees are distributed by age.
Most are between 25 and 40 years old.
The shape of the distribution is curved, indicating how age is distributed among the workforce.

```python
sns.histplot(df["Age"], bins=5, kde=True)
plt.title("Age Analysis")
plt.show
```

## ðŸ“Š Boxplot of salary by department
It shows the distribution of salary values for each department separately.
The IT department has *the highest median salary*.
The HR department shows *the smallest range*, while Sales and IT have a *wider spread*.
Some departments include *outliers*, indicating unusually high or low salaries.

```python
sns.boxplot(x="Department", y="Salary", data=df)
plt.title("Salary By Department")
plt.show
```

## ðŸ“Š Correlation matrix
It schows how closely the numeric columns in our data frame are related to each other. 
 **1** = perfect positive correlation (when one increases, the other also increases),
**-1** = perfect negative correlation (when one increases, the other decreases),
 **0** = no correlation.

A positive correlation between salary and age (r = 0.6), meaning older employees tend to earn more.
Correlations with the Left (employee turnover) column can help identify patterns linked to employee attrition.

```python


print(df.corr(numeric_only=True).round(2))
```

## ðŸ“Š Scatter plot
It helps visually examine the relationship between salary and length of employment in a company â€“ color-coded by whether they left the company or stayed.
Employees with lower salaries and shorter tenure are *more likely to leave*.
Employees with longer experience and higher salary are *more likely to stay*.

```python
sns.scatterplot(x="Length", y="Salary", hue="Left", data=df)
plt.title("Salary vs. Length of Employment")
plt.xlabel("Length of Employment (yrs)")
plt.ylabel("Salary (â‚¬)")
plt.show
```
